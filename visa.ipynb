```python
 import requests
```

```python
response = requests.get("https://visaguide.world/asia/")
```

```python
html_doc = response.text
```

```python
soup = BeautifulSoup(html_doc,"html.parser")
```

```python
tables =soup.find_all("table")
len(tables)
```

```python
headings = tables[1].find_all("th")
headings
```
